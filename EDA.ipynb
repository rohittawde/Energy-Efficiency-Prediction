## Data set and Attribute information 
This dataset consists of 768 observations with eight differentiating features. Given the dataset the goal is to predict two real valued responses for the heating and cooling load (energy efficency) using the given features. The features are summarised below :
$$
X1 	:Relative Compactness \\
X2	:Surface Area \\
X3	:Wall Area \\
X4	:Roof Area \\
X5	:Overall Height \\ 
X6	:Orientation \\
X7	:Glazing Area \\
X8	:Glazing Area Distribution \\ 
Y1	:Heating Load \\
Y2	:Cooling Load \\
$$

As we can observe from the description that all of our independant features are quantiative numeric variables. The goal of this project is to perform  predictive analytics that is make analyze current data to make predictions about the future using many techniques from machine learning.

```python
import pandas as pd
# Reading the data from a CSV file
heating = pd.read_csv("heating_1.csv")
cooling = pd.read_csv("cooling_1.csv")
heating_load = heating["Y1"]
cooling_load = cooling["Y2"]
X = heating.drop(['Y1'], axis = 1)
```

We split the given dataset from the csv format into a pandas dataframe. This dataframe is further split into dependant varibles viz. heating load and the cooling load and the independant variables.

```python
#Checking the data types of the series.
print("Data types of the independant features \n")
print(X.dtypes)
print("\nData types of the observations \n")
print("Heating load")
print(heating_load.dtype)
print("\nCooling load")
print(cooling_load.dtype)
```

## Visualization of the data

The Seaborn library in python is used to create effective visualizations of each independant feature with the corresponding response so as to create a scatter plot along with a histogram to show how each variable is distributed across it's range. This procedure is usefull for determining outliers which are the points that deviate significantly from the spread of the distribution.

Following are the scatter plots and the heat maps ot better visualize the correlation in between the variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes = True)
```

```python
sns.jointplot(X['X1'],heating_load,kind = 'scatter')
sns.jointplot(X['X2'],heating_load,kind = 'scatter')
sns.jointplot(X['X3'],heating_load,kind = 'scatter')
sns.jointplot(X['X4'],heating_load,kind = 'scatter')
sns.jointplot(X['X5'],heating_load,kind = 'scatter')
sns.jointplot(X['X6'],heating_load,kind = 'scatter')
sns.jointplot(X['X7'],heating_load,kind = 'scatter')
sns.jointplot(X['X8'],heating_load,kind = 'scatter')
plt.show()
```

```python
corr = heating.corr()
sns.heatmap(corr)
plt.show()
```
